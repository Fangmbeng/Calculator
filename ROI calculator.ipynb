```python
class calculator:

    def income(self, rental, laundary, miscelleanous):
        self.rental = rental
        self.laundary = laundary
        self.miscelleanous = miscelleanous

    def total_income(self):
        self.total_income = self.rental + self.laundary + self.miscelleanous
        return f"Total monthly income{self.total_income}"

    def expenses(self, tax, insurance, utility, HOA, law_care, vacancy, repairs, capEx, prop_management, mortgage):
        self.tax = tax
        self.insurance = insurance
        self.utility = utility
        self.HOA = HOA
        self.law_care = law_care
        self.vacancy = vacancy
        self.repairs = repairs
        self.capEx = capEx
        self.prop_management = prop_management
        self.mortgage = mortgage

    def total_expenses(self):
        self.total_expenses = self.tax + self.insurance + self.utility + self.HOA + self.law_care + self.vacancy + self.capEx + self.prop_management + self.mortgage
        return f"Total monthly expenses{self.total_expenses}"

    def cash_flow(self):
        self.cash_flow = self.total_income - self.total_expenses
        return f"total monthly cashflow {self.cashflow}"

    def cash_on_cash_ROI(self, down_payment, closing_cost, rehab_budget, others):
        self.down_payment = down_payment
        self.closing_cost = closing_cost
        self.rehab_budget = rehab_budget
        self.others = others

    def total_investment(self):
        self.total_investment = self.down_payment + self.closing_cost + self.rehab_budget + self.others

    def method_ROI(self):
        (self.cash_flow * 12) / self.total_investment * 100
        return

    def main():
        while True:
            action = input("Do you wish to calculate ROI or not. yes/no: ")
            if action.lower() != "no":
                print("If option is not applicable to you. Input 0")
                rental = int(input("Rental: "))
                tax = int(input("Tax: "))
                insurance = int(input("Insurance: "))
                utility = int(input("Utility: "))
                HOA = int(input("Home owners association fee: "))
                law_care = int(input("Law or snow care: "))
                vacancy = 0.05 * rental
                capEx = int(input("Capital expenditure: "))
                prop_management = 0.10 * rental
                mortgage = int(input("Mortgage: "))
                laundary = int(input("laundary income: "))
                miscelleanous = int(input("Miscelleanous: "))
                others = int(input("Other expenditure: "))
                down_payment = int(input("Down payments: "))
                closing_cost = int(input("Closing cost: "))
                rehab_budget = int(input("Rehab Budget: "))
                repairs = int(input("Repairs: "))

                total_income = rental + laundary + miscelleanous
                total_expenses = tax + insurance + utility + HOA + law_care + vacancy + capEx + prop_management + mortgage + repairs
                cash_flow = total_income - total_expenses
                total_investment = down_payment + closing_cost + rehab_budget + others
                cash_on_cash_ROI = ((cash_flow * 12) / total_investment) * 100

                print(f"INCOME"
                      f"Rental: $ {rental} " \
                      f"Laundary: $ {laundary} " \
                      f"Miscelleanous: $ {miscelleanous} " \
                      f"" \
                      f"Total Income : $ {total_income} " \
                      f"=============================" \
                      f"" \
                      f"EXPENSES" \
                      f"Tax: $ {tax} " \
                      f"Insurance: $ {insurance} " \
                      f"Utility: $ {utility} " \
                      f"Home Owners association: $ {HOA} " \
                      f"law or snow care: $ {law_care} " \
                      f"Vacancy: $ {vacancy} " \
                      f"Capital Expenditure: $ {capEx} " \
                      f"Property Management: $ {prop_management} " \
                      f"Mortgage: $ {mortgage} " \
                      f"" \
                      f"Total Expenses: $ {total_expenses} " \
                      f"=============================" \
                      f"" \
                      f"CASH FLOW" \
                      f"Total income: $ {total_income} " \
                      f"-" \
                      f"Total Expenses: $ {total_expenses} " \
                      f"" \
                      f"Monthly Cash flow: $ {cash_flow} " \
                      f"=============================" \
                      f"" \
                      f"TOTAL INVESTMENT" \
                      f"Down payment: ${down_payment} " \
                      f"Closing Cost: $ {closing_cost} " \
                      f"Rehab Budget: $ {rehab_budget} " \
                      f"Other Expenses: $ {others} " \
                      f"Cash on Cash Return on Investment: (Monthly CASH FLOW * 12) / TOTAL INVESTMENT" \
                      f"{cash_on_cash_ROI}% ")
                
                if cash_on_cash_ROI > 7 :
                    print("BiggerPockets remarks: Good financial Strategy")

                else:
                    print("Bigger Pockets recommend"
                          "1)House renovation to increase its value in future"
                          "2)Increase in house rent to catchup with high mortgage and tax rates in future with risk of "
                          "inflation "
                          "3)Reduction on non-profitable house expenditures"
                          "4)Would you like to contact a real estate agent")

            else:
                break

    main()

```
