```python
class calculator():


    def __init__(self, totalinvestment, totalincome, totalexpenses, cashflow, tax, insurance, utility, hOA, law_care,
                 vacancy, repairs, capEx, prop_management, mortgage, rental, laundary, miscelleanous, down_payment,
                 closing_cost, rehab_budget, others):
        self.totalincome = totalincome
        self.totalexpenses = totalexpenses
        self.cashflow = cashflow
        self.tax = tax
        self.insurance = insurance
        self.utility = utility
        self.hOA = hOA
        self.law_care = law_care
        self.vacancy = vacancy
        self.repairs = repairs
        self.capEx = capEx
        self.prop_management = prop_management
        self.mortgage = mortgage
        self.rental = rental
        self.laundary = laundary
        self.miscelleanous = miscelleanous
        self.down_payment = down_payment
        self.closing_cost = closing_cost
        self.rehab_budget = rehab_budget
        self.others = others
        self.totalinvestment = totalinvestment

    def total_income(self):
        self.totalincome = self.rental + self.laundary + self.miscelleanous
        return f"Total monthly income{self.totalincome}"

    def total_expenses(self):
        self.totalexpenses = self.tax + self.insurance + self.utility + self.hOA + self.law_care + self.vacancy + self.capEx + self.prop_management + self.mortgage
        return f"Total monthly expenses{self.totalexpenses}"

    def cash_flow(self):
        self.cashflow = self.totalincome - self.totalexpenses
        return f"total monthly cash flow {self.cashflow}"

    #
    # def cash_on_cash_ROI(self, down_payment, closing_cost, rehab_budget, others):
    #     self.down_payment = down_payment
    #     self.closing_cost = closing_cost
    #     self.rehab_budget = rehab_budget
    #     self.others = others

    def total_investment(self):
        self.totalinvestment = self.down_payment + self.closing_cost + self.rehab_budget + self.others
        return self.totalinvestment

    def method_ROI(self):
        return (self.cashflow * 12) / self.totalinvestment * 100

    def main():
        while True:
            action = input("Do you wish to calculate ROI or not. yes/no: ")
            if action.lower() != "no" or action.lower() == "yes":
                try:
                    print("If option is not applicable to you. Input 0")
                    rental = int(input("Rental: "))
                    tax = int(input("Tax: "))
                    insurance = int(input("Insurance: "))
                    utility = int(input("Utility: "))
                    hOA = int(input("Home owners association fee: "))
                    repairs = int(input("Repairs: "))
                    law_care = int(input("Law or snow care: "))
                    vacancy = 0.05 * rental
                    capEx = int(input("Capital expenditure: "))
                    prop_management = 0.10 * rental
                    mortgage = int(input("Mortgage: "))
                    laundary = int(input("laundary income: "))
                    miscelleanous = int(input("Miscelleanous: "))
                    others = int(input("Other expenditure: "))
                    down_payment = int(input("Down payments: "))
                    closing_cost = int(input("Closing cost: "))
                    rehab_budget = int(input("Rehab Budget: "))

                except:
                    return "Try Again"

                totalincome = rental + laundary + miscelleanous
                totalexpenses = tax + insurance + utility + hOA + law_care + vacancy + capEx + prop_management + mortgage + repairs
                cashflow = totalincome - totalexpenses
                totalinvestment = down_payment + closing_cost + rehab_budget + others
                cash_on_cash_ROI = ((cashflow * 12) / totalinvestment) * 100

                print(f"INCOME"
                      f"Rental: $ {rental}\n"
                      f"Laundary: $ {laundary}\n"
                      f"Miscelleanous: $ {miscelleanous}\n"
                      f"\n"
                      f"Total Income : $ {totalincome}\n"
                      f"=============================\n"
                      f"\n"
                      f"EXPENSES\n"
                      f"Tax: $ {tax}\n"
                      f"Insurance: $ {insurance}\n"
                      f"Utility: $ {utility} \n"
                      f"Home Owners association: $ {hOA}\n"
                      f"law or snow care: $ {law_care}\n"
                      f"Vacancy: $ {vacancy}\n"
                      f"Capital Expenditure: $ {capEx}\n"
                      f"Property Management: $ {prop_management}\n "
                      f"Mortgage: $ {mortgage}\n"
                      f"\n"
                      f"Total Expenses: $ {totalexpenses}\n"
                      f"=============================\n"
                      f"\n"
                      f"CASH FLOW\n"
                      f"Total income: $ {totalincome} \n"
                      f"        -\n"
                      f"Total Expenses: $ {totalexpenses}\n"
                      f"\n"
                      f"Monthly Cash flow: $ {cashflow} \n"
                      f"=============================\n"
                      f"\n"
                      f"TOTAL INVESTMENT\n"
                      f"Down payment: ${down_payment} \n"
                      f"Closing Cost: $ {closing_cost} \n"
                      f"Rehab Budget: $ {rehab_budget} \n"
                      f"Other Expenses: $ {others} \n"
                      f"Annual Cash on Cash Return on Investment: (Monthly CASH FLOW * 12 MONTHS) / TOTAL INVESTMENT\n"
                      f"{cash_on_cash_ROI}% ")

                if cash_on_cash_ROI > 7:
                    print("BiggerPockets remarks: Good financial Strategy")

                else:
                    print("Bigger Pockets recommend \n"
                          "1)House renovation to increase it's value in future \n"
                          "2)Increase in house rent to catchup with high mortgage and tax rates in future with risk of "
                          "inflation \n"
                          "4)Reduction on non-profitable house expenditures \n"
                          "5)Would you like to contact a real estate agent? \n")

            elif action.lower() == "no":
                break

    main()


cal = calculator
print(cal.main())
```
